```python

import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers

(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()


x_train, x_test = x_train / 255.0, x_test / 255.0


y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)

model_3_4 = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])


model_3_4.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])


history_3_4 = model_3_4.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))


plt.plot(history_3_4.history['accuracy'], label='Training accuracy')
plt.plot(history_3_4.history['val_accuracy'], label='Validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training and Validation Accuracy for Model 3.4')
plt.show()
```
